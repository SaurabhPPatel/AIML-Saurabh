```python
la = pd.read_csv('Laliga.csv',sep=',',header=1)
```

```python
la.replace(to_replace='-',value=0,inplace=True)
```

```python
la['Points'] = la['Points'].apply(pd.to_numeric)
```

```python
la[['Team','Points']].sort_values('Points',ascending=False).head(5)
```

```python
la['GoalsFor']=la['GoalsFor'].apply(pd.to_numeric)
la['GoalsAgainst']=la['GoalsAgainst'].apply(pd.to_numeric)
```

```python
def Goal_diff_count (df):
    df['Goal_diff_count']=df['GoalsFor']-df['GoalsAgainst']
    mx = df[['Team','Goal_diff_count']][df['Goal_diff_count']==df['Goal_diff_count'].max()]
    mn = df[['Team','Goal_diff_count']][df['Goal_diff_count']==df['Goal_diff_count'].min()]
    return(df[['Team','Goal_diff_count']],mx,mn)
```

```python
Goal_diff_count(la)
```

```python
la['GamesWon']=la['GamesWon'].apply(pd.to_numeric)
la['GamesPlayed']=la['GamesPlayed'].apply(pd.to_numeric)
```

```python
la['Winning Percent'] = 100*la['GamesWon']/la['GamesPlayed']
```

```python
la['Winning Percent'].isna()
```

```python
la['Winning Percent'].fillna(value=0,inplace=True)
```

```python
la.sort_values(by='Winning Percent',ascending = False).head(5)
```
